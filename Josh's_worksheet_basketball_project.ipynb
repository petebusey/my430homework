```
x = df_MiscStats['W'].corr(df_TeamStats['3P%'])  #correlation between wins and 3p%
y = df_MiscStats['W'].corr(df_MiscStats['eFG%']) #corelation between wins and eFG%
threes_taken = df_MiscStats['W'].corr(df_TeamStats['3PA']) #corelation between wins and 3PA
print("3P%", x)
print("eFG%", y)
print("3 att", threes_taken)
threes_made = df_MiscStats['W'].corr(df_TeamStats['3P'])
print("3 made", threes_made)
a = df_MiscStats['W'].corr(df_TeamStats['2P%'])
b = df_MiscStats['W'].corr(df_TeamStats['FG%'])
c = df_MiscStats['W'].corr(df_TeamStats['PTS'])
print("2P%", a)
print("FG%", b)
print("PTS", c)
```
